```python
nnumeric_cols = [feature for feature in dataset.columns if dataset[feature].dtypes != 'O']
nnumeric_cols
```

```python
for feature in nnumeric_cols:
    data_1 = dataset.copy()
    if 0 in dataset[feature].unique():
        pass
    else:
        data_1[feature] = np.log(data_1[feature])
        data_1.boxplot(column=feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
for feature in nnumeric_cols:
    median_value = dataset[feature].median()
    dataset[feature].fillna(median_value,inplace=True)
dataset[nnumeric_cols].isnull().sum()
```

```python
cat_col = [feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
cat_col
```

```python
def Replace_New_Feature(dataset,features_nan):
    data_1 = dataset.copy()
    data_1[features_nan] = data_1[features_nan].fillna("Missing")
    return data_1

dataset = Replace_New_Feature(dataset,cat_col)
dataset[cat_col].isnull().sum()
```

```python
dataset.describe(include=['object'])
```

```python
dataset.head(10)
```

```python
for feature in nnumeric_cols:
    data_1 = dataset.copy()
    data_1[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("count")
    plt.title(feature)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
for feature in cat_col:
    dataset[feature] = encoder.fit_transform(dataset[feature])
dataset.head(50)
```

```python
feature_scale =[feature for feature in dataset.columns if feature not in ['ID','year','Status']]
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
Scaler.fit(dataset[feature_scale])
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
y = dataset[['Status']]
```

```python
x = dataset.drop(['ID','Status'],axis=1)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state=42)
```

```python
x_train.shape, x_test.shape
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0))
feature_sel_model.fit(x_train,y_train)
```

```python
feature_sel_model.get_support()
```

```python
selected_feature = x_train.columns[feature_sel_model.get_support()]
```

```python
selected_feature
```

```python
print("Total feature:{}".format(x_train.shape[1]))
print("Selected feature:{}".format(len(selected_feature)))
print("feature with coefficients shrank to zero:{}".format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python
x_train[selected_feature].head()
```

```python
A_train = x_train[selected_feature]
```

```python
B_train = y_train
```

```python
A_train.shape,B_train.shape
```

```python
A_test = x_test[selected_feature]
```

```python
B_test = y_test
```

```python
A_test.shape,B_test.shape
```

```python
Parameter = {
    'criterion':['gini', 'entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
```

```python
cv = GridSearchCV(treemodel,param_grid=Parameter,cv=5,scoring='accuracy')
```

```python
cv.fit(A_train,B_train)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(A_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report 
```

```python
score=accuracy_score(y_pred,B_test)
```

```python
score
```

```python
print(classification_report(y_pred,B_test))
```
